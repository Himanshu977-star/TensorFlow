```python
import tensorflow as tf

# Disable eager execution to build static graph (TF1-style)
tf.compat.v1.disable_eager_execution()
```

```python
# Define the graph
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])
c = tf.add(a, b)
c
```

```python
tf.compat.v1.get_default_graph().as_graph_def()
```

```python
with tf.compat.v1.Session() as sess:
    writer = tf.compat.v1.summary.FileWriter(logdir='logs', graph=sess.graph)
    writer.close()
```

```python
with tf.compat.v1.Session() as sess:
    result = sess.run(c)
    print(result)
```

```python
%load_ext tensorboard
%tensorboard --logdir logs
```

🧾 Explanation of Each Line (Comments Separated from Code)
import tensorflow as tf
→ Imports the TensorFlow library for building and executing computation graphs.

tf.compat.v1.disable_eager_execution()
→ Disables TensorFlow 2.x's eager execution mode to allow building a static computation graph (like TensorFlow 1.x).

a = tf.constant([[1, 2], [3, 4]], name="a")
→ Creates a constant 2x2 matrix a as part of the computation graph.

b = tf.constant([[5, 6], [7, 8]], name="b")
→ Creates another 2x2 constant matrix b.

c = tf.matmul(a, b, name="matmul")
→ Adds a matrix multiplication operation to the graph: multiplies a and b.

tf.compat.v1.get_default_graph().as_graph_def()
→ Gets the full definition of the computation graph (optional; mainly for inspection/debugging).

with tf.compat.v1.Session() as sess:
→ Starts a session to initialize the graph and prepare for execution.

writer = tf.compat.v1.summary.FileWriter(logdir='logs', graph=sess.graph)
→ Writes the graph structure to the logs/ directory so it can be visualized in TensorBoard.

writer.close()
→ Closes the file writer to complete writing the log data.

with tf.compat.v1.Session() as sess:
→ Starts another session, this time to compute the matrix multiplication.

result = sess.run(c)
→ Executes the matmul operation defined in the graph and stores the result.

print(result)
→ Prints the matrix multiplication result: [[19, 22], [43, 50]].

%load_ext tensorboard
→ Loads the TensorBoard extension in a Jupyter Notebook environment.

%tensorboard --logdir logs
→ Launches TensorBoard in the notebook to visualize the graph saved in the logs folder.
